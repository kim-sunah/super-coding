### <b>2-7. 집합 자료형 (set)</b>

* 순서에 상관없이 중복을 제거하는 자료형
* 집합 자료형은 중복을 자동제거 한다
* 집합은 입력된 순서는 중요하지 않는다
* set() 생성자 함수 또는 {}를 이용해서 만들 수 있다
* 원소는 hashable 이어야 한다 (리스트는 원소로 추가할 수 없다)

#### <b>집합 만들기</b>

```python
set1 = {1, 2, 3, 3, 0}
print(set1, type(set1))

set2 = {3, 2, 1, ("a", "b"), ("a", "b")}
print(set2, type(set2))

set3 = set({1,2,3})
print(set3, type(set3))

set4 = set([1,2,3])
print(set4, type(set4))
```

#### <b>집합의 기본 메소드</b>

* <b>add(추가할 원소)</b> : 집합에 원소 추가
* <b>update(추가할 원소들)</b> : 집합에 여러 원소들 한번에 추가
* <b>remove(삭제할 원소)</b> : 집합에서 원소를 삭제
* <b>copy()</b> : 집합 데이터 복사
* <b>clear()</b> :  모든 집합 원소들 삭제

```python
set1 = set()
print(set1, type(set1))
for i in range(5):
    set1.add(i)
print(set1)

set1.add('a')
set1.add('문자열')
set1.add(34.6)
print(set1)

set1.update({11, 22, 33})
print(set1)

set1.remove(0)
set1.remove(2)
print(set1)

set2 = set1.copy()
print(set2, type(set2))

set2.clear()
print(set2)
```

```python
set1 = set({1, 2})
print(set1, type(set1))

# 튜플은 원소로 추가 할 수 있다
set1.add((1, 2))
print(set1)

# 리스트는 원소로 추가 할 수 없다
set1 = set()
set1.add([0, 1])
```

* 집합 자료형을 이용하여 리스트에서 중복 원소들 제거하기
  * 리스트의 순서가 중요하지 않을 경우
  * 리스트의 모든 원소가 hashable 일 경우

```python
list1 = [0, 1, 2, 3, 1, 5, 4, 0, 1, "a", "b", "a"]
list1 = list(set(list1))
print(list1, type(list1))
```

* 집합을 이용하여 중복 여부 확인

```python
nodes = [1, 2, 3, 1, 4, 2, 5]
visited = set()

for n in nodes:
  if n in visited:
    print("already visited", n)
  else:
    visited.add(n)
    print("new", n)
```

* 합집합, 교집합, 차집합

```python
set1 = {1,2,3,4,5,6}
set2 = {3,4,5,6,7,8}

# 합집합
set3 = set1 | set2
print("합집합:", set3)

set4 = set1.union(set2)
print("합집합:", set4)

# 교집합
set5 = set1 & set2
print("교집합:", set5)

set6 = set1.intersection(set2)
print("교집합:", set6)

# 차집합
set7 = set1 - set2
print("차집합:", set7)

set8 = set1.difference(set2)
print("차집합:", set8)

# *_update() method를 사용하면 원래의 set가 바뀐다. 결과 값은 None 이므로 주의해야 한다.
set9 = set1.difference_update(set2)
print("set1:", set1)
print("set9:", set9)
```
