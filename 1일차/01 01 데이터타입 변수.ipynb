## <b>데이터 타입</b>

파이썬에서 데이터 리터럴은 크게 다음과 같은 유형이 있다
* 숫자(Number) - int, float
* 문자열(Text) - str
* 논리값(Boolean) - bool
* 복합객체(Collectable) - list, dict, tuple, set

각각의 유형별로 자세히 아래에서 살펴본다

## <b>1. 변수(Variable)</b>
* 변수는 숫자나 문자 같은 <b>데이터 값을 저장하는 그릇</b>과 같다
* 변수의 선언 및 초기화는 변수의 값을 할당 할 때 이루어 진다
* 변수의 값은 등호 "="를 이용해서 할당한다

````python
year = 2025
month = 1
day = 31
print(year, month, day)
````

* 선언되지 않은 변수를 사용하면 에러가 발생한다

````python
print(birthday)
````

#### <b>변수의 이름 규칙</b>

* 변수이름은 알파벳(A-z), 숫자(0-9), 그리고 언더스코어('_')를 포함할 수 있다
  * 변수이름에 한글을 사용할 수는 있지만, 일반적으로 한글을 변수이름에 사용하는 것은 비추한다.
* 변수이름은 문자와 언더스코어로 시작할수 있다.
* 하지만, 변수이름은 숫자로는 시작할 수 없다
* 변수 이름은 대소문자 구분을 한다
* 파이썬의 키워드(if, then, import, def, 등 등)는 변수 이름으로 사용할 수 없다
* 선언된 변수의 이름은 바꿀 수 없다

#### <b>올바른 변수 이름 예시<b>


````python
myName = "John Smith"       # 카멜 표기법(Camel Case)
MyName = "John Smith"       # 파스칼 표기법(Pascal Case)
my_name = "John Smith"      # 스네이크 표기법(Snake Case)
_my_name = "Join Smith"
_myname = "John Smith"
myname = "John Smith"
MYNAME = "John Smith"
__myname__ = "john Smith"
myname2 = "John Smith"
my_2nd_name = "John Smith"
_ = "any data"
````

#### <b> 잘못된 변수 이름 예시</b>


````python
2nd_name = "John Smith"     #숫자로 시작
my-name = "John Smith"      #'-'포함
my name = "John Smith"      #중간에 스페이스
pass = "John Smith"         #pass는 파이썬 키워드
````

#### <b>변수에 데이터 리터럴 값 할당 하는 방법들</b>

````python
year = 2025
month = 5
day = 15
print(year, month, day)
````

````python
# 여러 변수값 한번에 초기화
year, month, day = 2025, 5, 15
print(year, month, day)
````

````python
# 다른 자료형 데이터를 한번에 설정
name, age, height = "John Smith", 24, 174.5
print(name, age, height)
````

````python
# 다른 변수의 값을 가져오기
data1 = 100
data2 = data1
print(data2, type(data2))
````

````python
# 계산 값을 변수에 할당 하기
x, y = 10, 20
result = x + y + 30
print(result)
````

````python
# 자기 값에 1을 더하기
x = 10
x = x + 10
print(x)

# 자기 값에 1을 빼기
x = x - 1
print(x)
````

````python
# 파이썬에서는 위의 표현을 다음과 같이 할 수 있다
x = 10
x += 10  # x = x + 1
print(x)

x -= 1  # x = x - 1
print(x)
````

````python
# 두개의 변수 값을 서로 바꾸기 (일반적인 방법)
a = 10
b = 20

_ = a
a = b
b = _
print(a, b)
````

````python
# 두개의 변수 값을 서로 바꾸기 (파이썬에서만 가능한 방법)
a = 10
b = 20

a, b = b, a
print(a, b)
````

#### <b>변수 값 출력 및 입력</b>

* print( ) : 변수 값 화면에 출력
* input( ) : 변수 값 키보드로 입력 받기

````python
# 변수 값 출력
name = "홍길동"
age = 24

print(name)
print(age)
print(name, age)
print("이름:", name, ", 나이:", age)
````

````python
# 변수 값 직접 입력 받기
name = input()
print(name, type(name))

age = int(input())
print(age, type(age))
````

````python
# 변수 값 입력받을때 정보 출력하고 입력받기
name = input("이름을 입력하세요:")
age = int(input("나이를 입력하세요:"))

print("이름:", name, ", 나이:", age)
````
