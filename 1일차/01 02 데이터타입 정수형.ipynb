### <b>2-1. 정수형 (int)</b>
* 양의 정수, 0, 음의 정수
* 소수점 이하 숫자는 없음
* 숫자의 크기에 제한이 없음 (메모리 한계 안에서)

```python
# 정수형
number1 = 1_234_567_890
number2 = 0
number3 = -1234567890
print(number1, type(number1))
print(number2, type(number2))
print(number3, type(number3))

bin_num = 0b10   #2진수
oct_num = 0o10   #8진수
dec_num = 10     #10진수
hex_num = 0x10   #16진수
print(bin_num, type(bin_num))
print(oct_num, type(oct_num))
print(dec_num, type(dec_num))
print(hex_num, type(hex_num))

number4 = 123_456_789             #3.6 버전 이후
print(number4, type(number4))
```

#### <b>정수형으로 변환하기</b>

```python
number1 = int("1234567890")      #10진수 스트링을 int로 변환
number2 = int("0x10", 16)        #16진수 스트링을 int로 변환
number3 = int(-1234567890.123)   #실수를 int 로 변환
print(number1, type(number1))
print(number2, type(number2))
print(number3, type(number3))
```

* 잘못된 정수의 형식을 int() 로 변환하려고 하면 에러가 난다

```python
# 계산식 문자열을 정수로 변환 할수 없다
int("123 + 11")
```

```python
# 부호 다음에 스페이스(빈칸)이 있으면 안된다
int("- 13")
```

```python
# 10 진수가 아닌 경우 base 를 지정해 주어야 한다
int("0b0101")
```

```python
# int 데이터가 아닌 문자열은 변환 할수 없다
int("123.456")
```

```python
# python 큰 정수형 데이터

x = 10000000000000000000000000000000000000000000;
print(x, type(x))

x = 10 ** 400
print(x, type(x))

x = x + 1
print(x)
```

```python

# #### <b>정수형 연습문제 2</b>
# * 사용자로부터 두 개의 정수를 입력받아 두 수의 합, 차, 곱, 몫, 나머지를 출력하는 프로그램을 작성하세요.
# * 몫은 // 연산자를 사용하여 정수 몫을 구하세요.

num1 = int(input("첫 번째 정수를 입력하세요: "))
num2 = int(input("두 번째 정수를 입력하세요: "))

print("두 수의 합:", num1 + num2)
print("두 수의 차:", num1 - num2)
print("두 수의 곱:", num1 * num2)
print("두 수의 몫:", num1 // num2)
print("두 수의 나머지:", num1 % num2)


# #### <b>정수형 연습문제 3</b>
# * 사용자로부터 정수를 하나 입력받아 그 수가 짝수인지 홀수인지 판별하는 프로그램을 작성하세요.

num = int(input("정수를 입력하세요: "))

if num % 2 == 0:
    print(num, "은(는) 짝수입니다.")
else:
    print(num, "은(는) 홀수입니다.")

```
